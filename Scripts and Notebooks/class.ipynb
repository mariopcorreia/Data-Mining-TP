```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold

# Carregar o conjunto de dados (substitua 'dataset.csv' pelo nome do seu arquivo CSV)
data = pd.read_csv("/home/antoniolmf/Documentos/md/tp/Data-Mining-TP/DatasetsOriginais/transfermarkt/transfermarkt_fbref_201718.csv", delimiter = ';',  index_col=0)

# Separar as colunas numéricas
numeric_columns = data.select_dtypes(include='number')

# Aplicar o limiar de variância
threshold = 20  # Defina o valor do limiar de variância desejado
selector = VarianceThreshold(threshold=threshold)
selected_columns = selector.fit_transform(numeric_columns)

# Obter os nomes das colunas selecionadas
selected_column_names = numeric_columns.columns[selector.get_support()].tolist()

# Calcular a variância e o limiar de variância para cada coluna selecionada
variance_thresholds = {}
for col in selected_column_names:
    variance = numeric_columns[col].var()
    variance_thresholds[col] = variance

# Exibir as colunas selecionadas e seus respectivos limiares de variância
print("Colunas selecionadas e seus limiares de variância:")
for col in selected_column_names:
    print(col, ":", variance_thresholds[col])
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Carregar o conjunto de dados usando pandas (supondo que seja um arquivo CSV)
data = pd.read_csv("/home/antoniolmf/Documentos/md/tp/Data-Mining-TP/DatasetsOriginais/transfermarkt/transfermarkt_fbref_201718.csv", delimiter = ';',  index_col=0)

# Escolher a coluna do valor de mercado dos jogadores
coluna_valor_mercado = 'value'

# Definir os limites dos intervalos (bins)
limites = [0, 5000000, 10000000, 20000000, 30000000, 40000000, 50000000, 100000000, 125000000, float('inf')]  # Defina os limites desejados para cada intervalo

# Definir os rótulos para os intervalos
rotulos = ['0-5 milhões', '5-10 milhões', '10-20 milhões', '20-30 milhões', '30-40 milhões' ,'40-50 milhões', '50-100 milhoes', '100 a 125', 'mais de 125']  # Defina os rótulos desejados para cada intervalo

# Criar a nova coluna com os intervalos binning
data['nova'] = pd.cut(data[coluna_valor_mercado], bins=limites, labels=rotulos, right=False)

# Converter os rótulos em valores numéricos usando LabelEncoder
label_encoder = LabelEncoder()
data['previsao'] = label_encoder.fit_transform(data['nova'])

# Imprimir o conjunto de dados resultante
#print(data)

data.head()

limite = 125e6  # Limite de valor

# Encontrar a linha que atende aos critérios
linha = data.loc[data[coluna_valor_mercado] > limite]

# Imprimir a linha encontrada
print(linha)
```
