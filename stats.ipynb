```python
import pandas as pd 
stats = pd.read_csv("DatasetsOriginais/stats.csv", delimiter = ';', encoding='Windows-1252')
```

```python
stats = stats.drop(['Rk', 'Age'],axis=1)
```

```python
stats = stats.rename(columns={'Player': 'PlayerName', 'Nation': 'nationality', 'Pos': 'Position', 'Squad': 'Affiliation', 'Comp': 'League'})
```

```python
def replace_position(position):
    if position.startswith("FW"):
        return "Forward"
    elif position.startswith("MF"):
        return "Midfielder"
    elif position.startswith("DF"):
        return "Defender"
    elif position.startswith("GK"):
        return "Goalkeeper"
    else:
        return position
    

stats["Position"] = stats["Position"].apply(replace_position)
```

```python
stats.head()
```

```python
stats.to_csv('stats_clean.csv', index=False)
```
