```python
# Carregando o dataset
players_df = pd.read_csv('DatasetsOriginais/players.csv', delimiter=',')

players_df.rename(columns={
    'name': 'PlayerName',
    'current_club_name': 'Affiliation',
    'height_in_cm' : 'Height (meters)',
    'country_of_citizenship' : 'Citizenship 1',
    'foot': 'Foot',
    'position': 'Position 1',
    'sub_position': 'Position 2',
    'market_value_in_eur': 'Market Value (Euros)',
    'highest_market_value_in_eur': 'Highest Market Value (Euros)',
    'agent_name': 'Agent',
    'contract_expiration_date': 'ContractExpiration',
    'current_club_domestic_competition_id': 'League',
    'country_of_birth' : 'nationality',
    'date_of_birth' : 'Age'
}, inplace=True)

# selecionando apenas as colunas necessárias
players_df = players_df[[
    'PlayerName', 'Affiliation',
    'Citizenship 1',
    'Position 1', 'Position 2', 'Foot', 'Height (meters)',
    'Market Value (Euros)', 'Highest Market Value (Euros)', 'Agent',
    'ContractExpiration', 'League', 'nationality', 'Age'
]]

# salvando em um novo arquivo csv
players_df.to_csv('players_transformed.csv', index=False)
```

```python
# Lê os datasets
competitions = pd.read_csv('DatasetsOriginais/competitions.csv')

# Cria um dicionário com o código do país e o nome por extenso
country_dict = dict(zip(competitions['competition_id'], competitions['name']))

# Atualiza a coluna "League" com os nomes por extenso
players_df['League'] = players_df['League'].map(country_dict)

# Salva o novo dataset
players_df.to_csv('players_transformed.csv', index=False)
```

```python
# Altera o nome da coluna Position para Position 1
players_df = players_df.rename(columns={'Position 1': 'Position'})

# Seleciona apenas as colunas em comum
common_columns = ['PlayerName', 'Affiliation', 'League', 'Height (meters)',
               'Citizenship 1', 'Age', 'Position', 'Position 2', 'Foot', 'Agent',
               'ContractExpiration', 'nationality', 'Market Value (Euros)', 'Highest Market Value (Euros)']

# Reordena as colunas do players_transformed.csv
players_df = players_df[common_columns]

# Salva o novo dataset
players_df.to_csv('players_transformed.csv', index=False)
```

```python
players_df['Agent'] = players_df['Agent'].fillna('No Agent')
```
